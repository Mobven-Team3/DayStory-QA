```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from reportlab.lib.utils import ImageReader
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics

# Load the CSV file and skip the first 12 rows
df = pd.read_csv('/Users/gorkemagkurt/Desktop/Stajlar/Mobven/proje/Daystory manuel test/daystory.csv', skiprows=12)

# Replace the placeholder column names with the actual column names from your CSV file
column_scenario_type = 'Scenerio Type'
column_test_id = 'Test ID'
column_module_section = 'Modül/Section/page'
column_case_title = 'Case Title'
column_priority = 'Priority'
column_test_objective = 'Test Objective'
column_device_info = 'Device İnfo/ \nBrowser info'
column_status = 'Status'  # New column for test results

# Set style for the plots
sns.set_style(style="whitegrid")

# Scenario Type Distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x=column_scenario_type)
plt.title('Senaryo Türü Dağılımı')
plt.savefig('scenario_type_distribution.png')
plt.close()

# Total Number of Tests
total_tests = df[column_test_id].nunique()

# Module/Section/Page Distribution
plt.figure(figsize=(15, 10))  # Grafik boyutunu artır
ax = sns.countplot(data=df, y=column_module_section, order=df[column_module_section].value_counts().index)
plt.title('Modül/Bölüm/Sayfa Dağılımı')
# Y ekseni etiketlerini döndürme ve font boyutunu küçültme
ax.set_yticklabels(ax.get_yticklabels(), rotation=45, ha='right', fontsize=8)
plt.tight_layout()  # Grafik öğelerinin sığmasını sağla
# Boş sütun adını doldurma
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    if width == 0:
        ax.text(x + width / 2, y + height / 2, '0', ha='center', va='center')

plt.savefig('module_distribution.png')
plt.close()




# Duplicate Case Titles
case_title_counts = df[column_case_title].value_counts()
duplicate_case_titles = case_title_counts[case_title_counts > 1]

# Priority Distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x=column_priority)
plt.title('Öncelik Dağılımı')
plt.savefig('priority_distribution.png')
plt.close()

# Duplicate Test Objectives
test_objective_counts = df[column_test_objective].value_counts()
duplicate_test_objectives = test_objective_counts[test_objective_counts > 1]

# Device/Browser Info Distribution
device_info_counts = df[column_device_info].value_counts()
plt.figure(figsize=(10, 5))
device_info_counts.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Cihaz/Tarayıcı Bilgisi Dağılımı')
plt.ylabel('')
plt.savefig('device_info_distribution.png')
plt.close()

# Status Distribution (Pie Chart)
status_counts = df[column_status].value_counts()
plt.figure(figsize=(10, 5))
status_counts.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Test Durumu Dağılımı')
plt.ylabel('')
plt.savefig('status_distribution.png')
plt.close()

# PDF Generation
pdf_file = "test_report.pdf"
c = canvas.Canvas(pdf_file, pagesize=letter)
width, height = letter

# Register the Arial Unicode MS font with ReportLab
pdfmetrics.registerFont(TTFont('ArialUnicodeMS', 'Arial Unicode.ttf'))

# Title
c.setFont("ArialUnicodeMS", 20)
c.drawCentredString(width / 2.0, height - 40, "Web Test Raporu")

# Introduction
c.setFont("ArialUnicodeMS", 12)
text = [
    "*** Bu rapor, DayStory Projesi - WEB mimarisi üzerinde gerçekleştirilen web", 
    "testlerinin detaylı bir analizini sunmaktadır.",
    "*** Test süreci ve sonuçlarıyla ilgili görüntülemeler ve açıklamalar sunar.",
    "*** Daystory, AI istemcilerini aktif olarak kullanan bir planlama", 
    "yardımcısıdır.",
    "Bu test raporu, Daystory'nin web uygulamasının çeşitli özelliklerinin",
    "test edilmesine dayanmaktadır.",
    "*** Rapor, test sürecinde karşılaşılan senaryoların tipine, testlerin", 
    "modül/sekme sayfasına",
    "ve önceliğine, yinelenen başlık ve hedeflerin dağılımına,", 
    "cihaz/tarayıcı bilgisinin dağılımına",
    "ve test durumlarının dağılımına odaklanmaktadır.",
    "*** Projemizin Android ve iOS kısımları için de ayrı test raporları hazırlanacaktır.",
    "*** Bu raporun oluşturulması için Python programlama dili ve çeşitli kütüphaneler kullanılmıştır.",
    "*** Grafiklerin oluşturulması ve raporun PDF formatında kaydedilmesi için", 
    "Matplotlib, Seaborn ve ReportLab kütüphaneleri kullanılmıştır.",
    "*** Raporun oluşturulma mantığı, önce veri setinin yüklenmesiyle başlar. Ardından,",
    "veri seti üzerinde çeşitli görselleştirmeler oluşturulur ve bu görselleştirmeler PDF raporunda yer alır.",
    "Son olarak, rapor PDF formatında kaydedilir ve son kullanıcıya sunulur.",
    "                                                                       ",
    "                                                                       ",
    "----------------------------------------------------------------------------------------------------------------",
    "JIRA ortamında ortaya çıkan sorunlardan ötürü Daystory ekibi kendi çözümlerini",
    "başarıyla oluşturmuş ve kendi çözümlerini yürürlüğe almıştır."
]
y_position = height - 60
for line in text:
    c.drawString(30, y_position, line)
    y_position -= 15
c.showPage()

# Scenario Type Analysis
c.setFont("ArialUnicodeMS", 16)
c.drawCentredString(width / 2.0, height - 40, "Senaryo Türü Analizi")
c.drawImage(ImageReader('scenario_type_distribution.png'), 30, height - 330, width - 60, 250)
c.setFont("ArialUnicodeMS", 12)
text = [
    "Bu grafik, pozitif ve negatif testlerin dağılımını göstermektedir.",
    "Başarı veya başarısızlık oranını anlamakta yardımcı olur.",
    "Pozitif testler, beklenen sonuçları doğrulamak için yürütülen testlerdir.",
    "Negatif testler, beklenmeyen sonuçları ortaya çıkarmak için yürütülen testlerdir.",
    "Bu analiz, test senaryolarının çeşitliliğini ve etkinliğini değerlendirmeye yardımcı olur."
]
y_position = height - 400
for line in text:
    c.drawString(30, y_position, line)
    y_position -= 15
c.showPage()


# Total Number of Tests
c.setFont("ArialUnicodeMS", 16)
c.drawCentredString(width / 2.0, height - 40, "Toplam Test Sayısı")
c.setFont("ArialUnicodeMS", 12)
text = [
    "* Bu bölümde, veri setinde gerçekleştirilen toplam test sayısı sunulmaktadır.",
    "* Toplam test sayısı, veri setindeki benzersiz test kimliklerinin sayısıdır.",
    "* Her bir test kimliği, bir test senaryosunu veya test durumunu temsil eder.",
    "* Toplam test sayısının bilinmesi, test kapsamının genişliğini ve test sürecinin", 
    "büyüklüğünü belirlemede önemlidir.",
    "* Daha fazla test sayısı, daha kapsamlı bir test sürecini işaret edebilir,", 
    "bu da ürünün daha sağlam olması için daha fazla test senaryosunun yürütüldüğü", 
    "anlamına gelir."
]
y_position = height - 100
for line in text:
    c.drawString(30, y_position, line)
    y_position -= 15
c.showPage()


# Module/Section/Page Distribution
c.setFont("ArialUnicodeMS", 16)
c.drawCentredString(width / 2.0, height - 40, "Modül/Bölüm/Sayfa Dağılımı")
c.drawImage(ImageReader('module_distribution.png'), 30, height - 330, width - 60, 250)
c.setFont("ArialUnicodeMS", 12)
text = [
    "• Bu bölümde, testlerin farklı modüller veya bölümler arasındaki dağılımı",
    "görselleştirilmektedir.",
    "• Her bir modül veya bölüm, bir uygulamanın veya web sitesinin belirli bir," 
    "bölümünü veya işlevselliğini temsil eder.",
    "• Bu grafik, hangi alanların daha sık test edildiğini belirlemeye yardımcı olur.",
    "• Bir modül veya bölümün daha sık test edilmesi, o bölümdeki potansiyel", 
    "hataların tespit edilme olasılığını artırır.",
    "• Testlerin dengeli bir şekilde dağıtılması, genel test kapsamının etkin ",
    "bir şekilde yönetildiğini gösterir ve ürünün her bir bölümünün sağlam olduğunu doğrular."
]
y_position = height - 410
for line in text:
    c.drawString(30, y_position, line)
    y_position -= 20
c.showPage()


# Duplicate Case Titles
c.setFont("ArialUnicodeMS", 16)
c.drawCentredString(width / 2.0, height - 40, "Yinelenen Test Başlıkları")
c.setFont("ArialUnicodeMS", 12)
y_position = height - 60
for title, count in duplicate_case_titles.items():
    c.drawString(30, y_position, f"{title}: {count} kez")
    y_position -= 15
    if y_position < 40:  # New page if necessary
        c.showPage()
        c.setFont("ArialUnicodeMS", 16)
        c.drawCentredString(width / 2.0, height - 40, "Yinelenen Test Başlıkları")
        c.setFont("ArialUnicodeMS", 12)
        y_position = height - 60
c.showPage()

# Priority Distribution
c.setFont("ArialUnicodeMS", 16)
c.drawCentredString(width / 2.0, height - 40, "Öncelik Dağılımı")
c.drawImage(ImageReader('priority_distribution.png'), 30, height - 330, width - 60,250)
c.setFont("ArialUnicodeMS", 12)
c.drawString(30, height - 350, "Bu grafik, test vakalarının öncelik seviyelerine göre dağılımını gösterir.")
c.drawString(30, height - 365, "Test vakalarının kritikliğini anlamaya yardımcı olur.")
c.showPage()

# Duplicate Test Objectives
c.setFont("ArialUnicodeMS", 16)
c.drawCentredString(width / 2.0, height - 40, "Yinelenen case Objective'leri")
c.setFont("ArialUnicodeMS", 12)
y_position = height - 60
for objective, count in duplicate_test_objectives.items():
    c.drawString(30, y_position, f"{objective}: {count} kez")
    y_position -= 15
    if y_position < 40:  # New page if necessary
        c.showPage()
        c.setFont("ArialUnicodeMS", 16)
        c.drawCentredString(width / 2.0, height - 40, "Yinelenen case Objective'leri")
        c.setFont("ArialUnicodeMS", 12)
        y_position = height - 60
c.showPage()

# Device/Browser Info Distribution
c.setFont("ArialUnicodeMS", 16)
c.drawCentredString(width / 2.0, height - 40, "Cihaz/Tarayıcı Bilgisi Dağılımı")
c.drawImage(ImageReader('device_info_distribution.png'), 30, height - 330, width - 60, 250)
c.setFont("ArialUnicodeMS", 12)
c.drawString(30, height - 350, "Bu grafik, test vakalarının farklı cihazlar veya tarayıcılara göre dağılımını gösterir.")
c.drawString(30, height - 365, "Çeşitli platformlarda test kapsamını anlamaya yardımcı olur.")
c.showPage()

# Test Status Distribution
c.setFont("ArialUnicodeMS", 16)
c.drawCentredString(width / 2.0, height - 40, "Test Durumu Dağılımı")
c.drawImage(ImageReader('status_distribution.png'), 30, height - 330, width - 60, 250)
c.setFont("ArialUnicodeMS", 12)
c.drawString(30, height - 350, "Bu grafik, test sonuçlarının farklı durum kategorilerine göre dağılımını gösterir.")
c.drawString(30, height - 365, "Testlerin genel sonuçlarını anlamaya yardımcı olur.")
c.showPage()

# Conclusion
c.setFont("ArialUnicodeMS", 16)
c.drawCentredString(width / 2.0, height - 40, "Sonuç")
c.setFont("ArialUnicodeMS", 12)
text = [
    "Sonuç:",
    "Bu rapor, test vakalarının ve dağılımlarının bir özetini sunar.",
    "Görüntülemeler ve açıklamalar, test sürecini ve sonuçlarını derinlemesine anlamaya yardımcı olur."
]
y_position = height - 60
for line in text:
    c.drawString(30, y_position, line)
    y_position -= 15

# Prepared by
c.setFont("ArialUnicodeMS", 12)
c.drawString(10, 80, "Bu rapor; Python kodu ile otomatize edilmiştir ve") 
c.drawString(10, 65, "insan eli değmeden kod tarafından csv dosyasından data çekilerek yazılmıştır.")
c.drawString(10,50, "Hazırlayan: daystory QA.")

# Thanks to
c.setFont("ArialUnicodeMS", 12)
#c.drawString(10, 25, "Tüm daystory ekibine; Ayrıca, QA ekibinden Gülşen ve Zeynep'e ")
#c.drawString(10, 5, "dünyanın en destekleyici takım arkadaşları oldukları için teşekkürler!")

# Save the PDF
c.save()

print("Rapor başarıyla oluşturuldu ve 'test_report.pdf' olarak kaydedildi.")

```
